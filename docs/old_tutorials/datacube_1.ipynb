```python
import matplotlib.pyplot as plt
from eddy import rotationmap
import numpy as np
```

```python
cube = rotationmap('../../cubes/TWHya_CO_cube_v0.fits', FOV=3.0, downsample='beam')
cube.plot_data()
```

```python
params = {}
params['x0'] = 0
params['y0'] = 1
params['PA'] = 2
params['mstar'] = 3
params['vlsr'] = 4

params['inc'] = 6.5
params['dist'] = 60.1

p0 = [0.0, 0.0, 151.0, 0.8, cube.vlsr]

cube.fit_map(p0, params, nburnin=1000, nsteps=1000, optimize=False)
```
